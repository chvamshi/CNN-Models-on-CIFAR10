```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH = "/content/drive/MyDrive/Colab Notebooks/Saved_CSV/"

lenet5 = pd.read_csv(PATH+"lenet5.csv")
vgg11 = pd.read_csv(PATH+"vgg11.csv")
alexnet = pd.read_csv(PATH+"alex.csv")
googlenet = pd.read_csv(PATH+"googlenet.csv")
inceptionv3 = pd.read_csv(PATH+"inceptionv3.csv")
resnet50 = pd.read_csv(PATH+"resnet50.csv")
```

```python
def clean(df):
  df = df[['Train Loss', 'Train Accracy', 'Validation Loss', 'Validation Accuracy']]

  return df
```

```python
lenet5 = clean(lenet5)
vgg11 = clean(vgg11)
alexnet = clean(alexnet)
googlenet = clean(googlenet)
inceptionv3 = clean(inceptionv3)
resnet50 = clean(resnet50)
```

```python
sizes = [
        lenet5.shape[0],
        vgg11.shape[0],
        alexnet.shape[0],
        googlenet.shape[0],
        inceptionv3.shape[0],
        resnet50.shape[0]
]

max_size = max(sizes)
max_size
```

```python
# Train loss
EPOCHS = [i for i in range(max_size)]
plt.plot(EPOCHS, lenet5['Train Loss'], label = "Lenet5")
plt.plot(EPOCHS[:vgg11.shape[0]], vgg11['Train Loss'], label="VGG 11")
plt.plot(EPOCHS[:alexnet.shape[0]], alexnet['Train Loss'], label = "Alexnet")
plt.plot(EPOCHS[:googlenet.shape[0]], googlenet['Train Loss'], label = "Googlenet")
plt.plot(EPOCHS[:inceptionv3.shape[0]], inceptionv3['Train Loss'], label = "Inception v3")
plt.plot(EPOCHS[:resnet50.shape[0]], resnet50['Train Loss'], label = "Resnet50")
plt.legend()
plt.show()
```

```python
# Train Accuracy
plt.plot(EPOCHS, lenet5['Train Accracy'], label = "Lenet5")
plt.plot(EPOCHS[:vgg11.shape[0]], vgg11['Train Accracy'], label="VGG 11")
plt.plot(EPOCHS[:alexnet.shape[0]], alexnet['Train Accracy'], label = "Alexnet")
plt.plot(EPOCHS[:googlenet.shape[0]], googlenet['Train Accracy'], label = "Googlenet")
plt.plot(EPOCHS[:inceptionv3.shape[0]], inceptionv3['Train Accracy'], label = "Inception v3")
plt.plot(EPOCHS[:resnet50.shape[0]], resnet50['Train Accracy'], label = "Resnet50")
plt.legend()
plt.show()
```

```python
# Validation Loss
EPOCHS = [i for i in range(max_size)]
plt.plot(EPOCHS, lenet5['Validation Loss'], label = "Lenet5")
plt.plot(EPOCHS[:vgg11.shape[0]], vgg11['Validation Loss'], label="VGG 11")
plt.plot(EPOCHS[:alexnet.shape[0]], alexnet['Validation Loss'], label = "Alexnet")
plt.plot(EPOCHS[:googlenet.shape[0]], googlenet['Validation Loss'], label = "Googlenet")
plt.plot(EPOCHS[:inceptionv3.shape[0]], inceptionv3['Validation Loss'], label = "Inception v3")
plt.plot(EPOCHS[:resnet50.shape[0]], resnet50['Validation Loss'], label = "Resnet50")
plt.legend()
plt.show()
```

```python
# Validation Accuracy
multiplied_list1 = [element * 100 for element in lenet5['Validation Accuracy']]
multiplied_list2 = [element * 100 for element in vgg11['Validation Accuracy']]
multiplied_list3 = [element * 100 for element in alexnet['Validation Accuracy']]
multiplied_list4 = [element * 100 for element in googlenet['Validation Accuracy']]
multiplied_list5 = [element * 100 for element in inceptionv3['Validation Accuracy']]
multiplied_list6 = [element * 100 for element in resnet50['Validation Accuracy']]

EPOCHS = [i for i in range(max_size)]
plt.plot(EPOCHS, multiplied_list1 , label = "Lenet5")
plt.plot(EPOCHS[:vgg11.shape[0]], multiplied_list2 , label="VGG 11")
plt.plot(EPOCHS[:alexnet.shape[0]], multiplied_list3 , label = "Alexnet")
plt.plot(EPOCHS[:googlenet.shape[0]], multiplied_list4 , label = "Googlenet")
plt.plot(EPOCHS[:inceptionv3.shape[0]], multiplied_list5 , label = "Inception v3")
plt.plot(EPOCHS[:resnet50.shape[0]], multiplied_list6 , label = "Resnet50")
plt.legend()
plt.show()
```
